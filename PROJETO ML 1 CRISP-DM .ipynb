## Simulando a exploração e visualização de um conjunto de dados de vendas de varejo
### Limpeza e preparação dos dados

```python
# 1. Importação das Bibliotecas e Carregamento dos Dados

import pandas as pd

# Carregar os dados do arquivo CSV
df = pd.read_csv('Online Retail.csv')
df.head(20)
```

```python
# informações sobre os dados
df.info()
```

```python
# Verificar se há valores faltantes
# Verificamos a quantidade de valores faltantes em cada coluna para entender onde precisamos focar nossos esforços de limpeza.
df.isnull().sum()
```

```python
# verificar o tipo de uma coluna
df['CustomerID'].dtype
```

```python
# verificar o tipo de uma coluna
df['Description'].dtype
```

```python
# Converter 'InvoiceDate' para datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# identificar as entradas com quantidade negativa
df[df['Quantity'] < 0].head()
```

```python
# Remover entradas com quantidades negativas
df = df.loc[df['Quantity'] > 0]
```
